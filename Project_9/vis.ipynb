```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesDataAnalyzer:
    def __init__(self):
        self.data = None
        self.last_fig = None

    def load_data(self, file_path):
        try:
            self.data = pd.read_csv(file_path)
            print("\nDataset Loaded Successfully!\n")
        except Exception as e:
            print(f"\nError Loading File: {e} 🚫\n")

    def explore_data(self):
        if self.data is None:
            print("\nPlease Load The Dataset First. 🚫\n")
            return
            
        while True:
            print("\n================ Explore Data ================")
            print("1. Display The First 5 Rows")
            print("2. Display The Last 5 Rows")
            print("3. Display Column Names")
            print("4. Display Data Types")
            print("5. Display Basic Info")
            print("0. Go Back")
            print("==============================================")
            
            sub_choice = int(input("\nEnter Choice: "))
            
            match sub_choice:
                case 1:
                    print(f"\n{self.data.head()}\n")
                case 2:
                    print(f"\n{self.data.tail()}\n")
                case 3:
                    print(f"\n{self.data.columns.tolist()}\n")
                case 4:
                    print(f"\n{self.data.dtypes}\n")
                case 5:
                    print("\n")
                    self.data.info()
                    print("\n")
                case 0: 
                    break
                case _:
                    print("\nInvalid Choice. 🚫\n")

    def dataframe_operations(self):
        if self.data is None:
            print("\nPlease Load The Dataset First. 🚫\n")
            return
            
        while True:
            print("\n==== C. Search, Sort, Filter & Aggregate ====")
            print("1. Search For Specific Products")
            print("2. Sort Data Based On Sales")
            print("3. Filter Data For Specific Regions")
            print("4. Apply Aggregating Functions")
            print("0. Go Back")
            print("==============================================")
            
            sub_choice = int(input("\nEnter Choice: "))
            
            match sub_choice:
                case 1:
                    term = input("Enter Product Name To Search: ")
                    result = self.data[self.data['Product'].str.contains(term, case=False, na=False)]
                    print(f"\n{result}\n")
                case 2:
                    sorted_df = self.data.sort_values(by='Sales', ascending=False)
                    print(f"\n{sorted_df.head()}\n")
                case 3:
                    region = input("Enter Region To Filter (E.g. North): ")
                    filtered_df = self.data.query(f"Region.str.lower() == '{region.lower()}'", engine='python')
                    print(f"\n{filtered_df}\n")
                case 4:
                    agg_df = self.data.groupby('Region')['Sales'].agg(['sum', 'mean', 'count'])
                    print(f"\n{agg_df}\n")
                case 0: 
                    break
                case _:
                    print("\nInvalid Choice. 🚫\n")
    
    def handle_data(self):
        if self.data is None:
            print("\nPlease Load The Dataset First. 🚫\n")
            return
            
        while True:
            print("\n============ Handle Missing Data ============")
            print("1. Display Rows With Missing Values")
            print("2. Fill Missing Values With Mean")
            print("3. Drop Rows With Missing Values")
            print("4. Replace Missing Values With A Specific Value")
            print("0. Go Back")
            print("==============================================")
            
            sub_choice = int(input("\nEnter Choice: "))
            
            match sub_choice:
                case 1:
                    missing = self.data.isna().sum().sum()
                    print(f"\nTotal Missing Values: {missing}\n")
                case 2:
                    numeric_mean = self.data.mean(numeric_only=True)
                    self.data.fillna(numeric_mean, inplace=True)
                    print("\nMissing Numeric Fields Filled With Column Averages.\n")
                case 3:
                    self.data.dropna(inplace=True)
                    print("\nRows With Missing Values Dropped.\n")
                case 4:
                    replacement = input("\nEnter Replacement Value: ")
                    self.data.fillna(replacement, inplace=True)
                    print("\nValues Replaced.\n")
                case 0: 
                    break
                case _:
                    print("\nInvalid Choice. 🚫\n")

    def statistics(self):
        if self.data is None:
            print("\nPlease Load The Dataset First. 🚫\n")
            return
            
        while True:
            print("\n====== Generate Descriptive Statistics ======")
            print("1. Basic Description (Describe)")
            print("2. Standard Deviation & Variance")
            print("3. Quantiles")
            print("0. Go Back")
            print("==============================================")
            
            sub_choice = int(input("\nEnter Choice: "))
            
            match sub_choice:
                case 1:
                    print(f"\n{self.data.describe()}\n")
                case 2:
                    stats = self.data.select_dtypes(include=np.number).agg(['var', 'std'])
                    print(f"\n{stats}\n")
                case 3:
                    quantiles = self.data.quantile([0.25, 0.75], numeric_only=True)
                    print(f"\n{quantiles}\n")
                case 0:
                    break
                case _:
                    print("\nInvalid Choice. 🚫\n")

    def visualize_data(self):
        if self.data is None:
            print("\nPlease Load The Dataset First. 🚫\n")
            return
            
        while True:
            print("\n============= Data Visualization =============")
            print("1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Histogram")
            print("5. Pie Chart")
            print("0. Go Back")
            print("==============================================")
            
            try:
                sub_choice = int(input("\nEnter Choice: "))
            except ValueError:
                print("\nInvalid Input. Please Enter A Number. 🚫\n")
                continue
                
            if sub_choice == 0:
                break
                
            plt.figure(figsize=(7, 4))
            
            match sub_choice:
                case 1:
                    sales_sum = self.data.groupby('Region')['Sales'].sum()
                    sales_sum.plot(kind='bar', color='skyblue', title="Sales By Region")
                    plt.xlabel("Region")
                    plt.ylabel("Total Sales")
                case 2:
                    sales_trend = self.data.groupby('Year')['Sales'].sum()
                    sales_trend.plot(kind='line', marker='o', color='g', title="Sales Over Years")
                    plt.xlabel("Year")
                    plt.ylabel("Total Sales")
                case 3:
                    x_axis = input("X-Axis Column: ")
                    y_axis = input("Y-Axis Column: ")
                    if x_axis in self.data.columns and y_axis in self.data.columns:
                        sns.scatterplot(data=self.data, x=x_axis, y=y_axis, hue='Region')
                        plt.title(f"{y_axis} Vs {x_axis}")
                    else:
                        print("\nInvalid Columns. 🚫\n")
                        plt.close()
                        continue
                case 4:
                    if 'Sales' in self.data.columns:
                        self.data['Sales'].plot(kind='hist', bins=10, edgecolor='black', title="Sales Spread")
                        plt.xlabel("Sales Amount")
                        plt.ylabel("Frequency")
                    else:
                        print("\nSales Column Missing. 🚫\n")
                        plt.close()
                        continue
                case 5:
                    if 'Region' in self.data.columns and 'Sales' in self.data.columns:
                        sales_pie = self.data.groupby('Region')['Sales'].sum()
                        plt.pie(sales_pie, labels=sales_pie.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
                        plt.title("Sales Share By Region")
                    else:
                        print("\nRequired Columns Missing. 🚫\n")
                        plt.close()
                        continue
                case _:
                    print("\nInvalid Choice. 🚫\n")
                    plt.close()
                    continue

            self.last_fig = plt.gcf()
            plt.tight_layout()
            print("\nGenerating Plot...\n")
            plt.show()

    def save_visualization(self):
        if self.last_fig is None:
            print("\nNo Visualization Found! Please Generate A Plot In Option 6 First. 🚫\n")
            return
            
        print("\n============= Save Visualization =============")
        filename = input("\nEnter Filename (E.g., Plot.png): ")
        try:
            self.last_fig.savefig(filename, bbox_inches='tight')
            print(f"\nVisualization Saved Successfully As {filename}!\n")
        except Exception as e:
            print(f"\nFailed To Save: {e} 🚫\n")

def main():
    analyzer = SalesDataAnalyzer()
    
    while True:
        print("\n========== Data Analysis & Visualization Program ==========")
        print("1. Load Dataset")
        print("2. Explore Data")
        print("3. Perform DataFrame Operations")
        print("4. Handle Missing Data")
        print("5. Generate Descriptive Statistics")
        print("6. Data Visualization")
        print("7. Save Visualization")
        print("0. Exit")
        print("===========================================================")
        
        try:
            choice = int(input("\nEnter Your Choice: "))
        except ValueError:
            print("\nInvalid Input. Please Enter A Number. 🚫\n")
            continue
        
        match choice:
            case 1:
                path = input("\nEnter CSV Path: ")
                analyzer.load_data(path)
            case 2:
                analyzer.explore_data()
            case 3:
                analyzer.dataframe_operations()
            case 4:
                analyzer.handle_data()
            case 5:
                analyzer.statistics()
            case 6:
                analyzer.visualize_data()
            case 7:
                analyzer.save_visualization()
            case 0:
                print("\nExiting The Program. Goodbye! 👋👋\n")
                break
            case _:
                print("\nInvalid Choice, Please Try Again. 🚫\n")

if __name__ == "__main__":
    main()
```
