```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv('Air Quality.csv')

display(df.head())
```

```python
if 'Unnamed: 15' in df.columns:
    df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'])

df = df.dropna(how='all')
df = df.replace(-200, np.nan)

if 'NMHC(GT)' in df.columns:
    df = df.drop(columns=['NMHC(GT)'])

df = df.fillna(df.median(numeric_only=True))
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')
df = df.drop(columns=['Date', 'Time'])
df.set_index('Datetime', inplace=True)

print("Data Cleaning Complete. Missing values remaining:")
print(df.isnull().sum())
```

```python
weekly_data = df.resample('W').mean()

plt.figure(figsize=(14, 6))
plt.plot(weekly_data.index, weekly_data['CO(GT)'], color='crimson', marker='o', linestyle='-')
plt.title('Weekly Average Carbon Monoxide (CO) Levels Over Time')
plt.xlabel('Date')
plt.ylabel('CO Level')
plt.show()
```

```python
key_vars = ['CO(GT)', 'NOx(GT)', 'NO2(GT)', 'T', 'RH', 'AH']
display(df[key_vars].describe())
corr_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Air Quality and Weather Metrics')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(weekly_data.index, weekly_data['NOx(GT)'], color='indigo', marker='s', linestyle='-')
plt.title('Weekly Average Nitrogen Oxides (NOx) Levels Over Time')
plt.xlabel('Date')
plt.ylabel('NOx Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='T', y='CO(GT)', alpha=0.3, color='darkorange')
plt.title('Relationship Between Temperature and Carbon Monoxide (CO)')
plt.xlabel('Temperature (C)')
plt.ylabel('Carbon Monoxide (CO) Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='RH', y='NO2(GT)', alpha=0.3, color='teal')
plt.title('Relationship Between Relative Humidity and Nitrogen Dioxide (NO2)')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Nitrogen Dioxide (NO2) Level')
plt.show()
```
