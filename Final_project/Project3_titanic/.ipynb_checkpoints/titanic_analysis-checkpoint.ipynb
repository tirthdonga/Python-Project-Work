```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
df = pd.read_csv('test.csv')
df['Survived'] = df['Sex'].apply(lambda x: 1 if x == 'female' else 0)

display(df.head())
display(df.info())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
display(df[numerical_cols].describe())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Survived', hue='Sex', palette='pastel')
plt.title('Survival Count by Gender (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count of Passengers')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Count of Passengers')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Overall Age Distribution')
axes[0].set_xlabel('Age')
bins = [0, 12, 18, 60, 100]
labels = ['Child (0-12)', 'Teen (13-18)', 'Adult (19-60)', 'Senior (60+)']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

sns.countplot(data=df, x='AgeGroup', hue='Survived', palette='coolwarm', ax=axes[1])
axes[1].set_title('Survival Count by Age Group')
axes[1].set_xlabel('Age Group')

plt.tight_layout()
plt.show()
```
